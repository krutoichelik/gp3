```python
import glob
import os
import pandas as pd
```

```python
files = glob.glob('/content/*.txt') # ф-ия glob ищет все файлы с txt (чтоб отдельно все не считывать (тк много))

dfs = [] # все ссылки сюда закидываем
for x in files:
    df1 = pd.read_csv(x, header=None, names=['url'])
    dfs.append(df1)

df1.head(2)
```

удалим первую строку в каждом датасете, так как там не ссылка на сам рестик

```python
dfs = [df[df['url'].str.contains('/maps/org/')] for df in dfs]
dfs
```

```python
df_all = pd.concat(dfs, ignore_index=True)
df_all.shape
```

```python
df_all.head(2)
```

добавим столбец с id (вытащим из ссылки)

```python
from urllib.parse import urlparse

def id(url):
    return urlparse(url).path.split('/')[-2]  # извлекаем id

df_all['id'] = df_all['url'].apply(id) # применяем ко всем строкам

df_all
```

```python
uni_df = df_all.drop_duplicates(subset='id')
uni_df.shape
```

```python
df_1 = uni_df.iloc[:len(uni_df) // 2]
df_2 = uni_df.iloc[len(uni_df) // 2:]

print(df_1.shape)
print(df_2.shape)
```

```python
df_1.to_csv('final_df_links_part1.csv', index=False)
df_2.to_csv('final_df_links_part2.csv', index=False)
```
