```python
from selenium.webdriver.common.actions.wheel_input import ScrollOrigin
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
import time
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument("user-agent=МАКСИМКОРОЛЬПАРСИНГА")
url = "https://yandex.ru/maps/213/moscow/search/%D0%A0%D0%B5%D1%81%D1%82%D0%BE%D1%80%D0%B0%D0%BD/?filter=alternate_vertical%3ARequestWindow&ll=37.615835%2C55.725219&z=12"
driver = webdriver.Chrome(options=options)

try:
    driver.get(url=url)
    time.sleep(2)

    elem = driver.find_element(By.CSS_SELECTOR, "a.link-overlay")
    ActionChains(driver).move_to_element(elem).perform()

    origin = ScrollOrigin.from_element(elem)
    actions = ActionChains(driver)
    actions.scroll_from_origin(origin, 0, 400).perform()

    links = set()
    
    no_changes = 0
    last_href = None   # <-- вместо ref/last-объектов

    while no_changes != 50:
        elems = driver.find_elements(By.CSS_SELECTOR, "a.link-overlay")

        for el in elems:
            href = el.get_attribute("href")
            if href:
                links.add(href)

        print(len(links))

        if elems:
            last = elems[-1]
            current_href = last.get_attribute("href")   # <-- строка

            if current_href == last_href:
                no_changes += 1
            else:
                no_changes = 0

            last_href = current_href

            origin = ScrollOrigin.from_element(last)
            actions = ActionChains(driver)
            actions.scroll_from_origin(origin, 0, 400).perform()

        time.sleep(10)   # чуть больше, чтобы Яндекс успевал подгружать

        with open("yandex_restaurants_links.txt", "w", encoding="utf-8") as f:
            for h in sorted(links):
                f.write(h + "\n")

    print(links)

except Exception as ex:
    print(ex)

finally:
    driver.close()
    driver.quit()
```
